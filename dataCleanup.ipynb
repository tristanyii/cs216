```python
import pandas as pd
```

```python
df = pd.read_csv("output.csv")

df = df[df["last_price_dollars"].ne(0) & df["last_price_dollars"].notna()]
df = df[df["result"].notna()]
df = df[df["close_time"].notna()]

df["date"] = df["close_time"].str.extract(r"(^\d{4}-\d{2}-\d{2})")
df["date"] = pd.to_datetime(df["date"], format="%Y-%m-%d")
df = df.drop(columns="close_time")

df.to_csv("kalshi_data_clean.csv", index=False)
```
